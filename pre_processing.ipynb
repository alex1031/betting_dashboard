```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
df = pd.read_csv('basquete.csv')
```

```python
old_columns = df.columns
```

```python
df.rename(columns={'Data':'Date', 'Liga':'League', 'Entrada':'Entry', 'Fundamento':'Type',
                  'Investimento':'Investment', 'Odd ':'Odds', 'Ganho':'Gain', 'Resultado':'Result'}, inplace=True)
```

```python
df.head()
```

```python
# Only want data from NBA
df = df[df['League']=='NBA']
```

```python
df = df.reset_index().drop(columns='index')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
set(df['Type'])
```

```python
df.replace('1췈 Quarto', 'Points', inplace=True)
df.replace('3 Pontos', '3 Pointers', inplace=True)
df.replace('3 pontos', '3 Pointers', inplace=True)
df.replace('Assistencias', 'Assists', inplace=True)
df.replace('Assist칡ncias', 'Assists', inplace=True)
df.replace('Dupla', 'Pair', inplace=True)
df.replace('Pontos', 'Points', inplace=True)
df.replace('N칚o marcar', 'Do Not Score', inplace=True)
df.replace('Nenhum a marcar', 'Both to Not Score', inplace=True)
df.replace('Pts + Reb   ', 'Pts + Reb', inplace=True)
df.replace('Pts + Reb    ', 'Pts + Reb', inplace=True)
df.replace('Rebotes', 'Rebounds', inplace=True)
df.replace('Roubadas + Tocos', 'Stl + Blk', inplace=True)
df.replace('Roubos', 'Steals', inplace=True)
df.replace('Tocos', 'Blocks', inplace=True)
df.replace('Tripla', 'Triple', inplace=True)
df.replace('Vencedor', 'ML', inplace=True)
df.replace('Time Total', 'Total', inplace=True)
```
